```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("sonaralldata.csv", names = range(0,61), header= 0)
df
```

```python
# we can use PCA to limit the number of columns it will take the mean of giving number of columns and add them..
```

```python
x = df.iloc[:, 0: -1]
x
```

```python
x.shape
```

```python
from sklearn.decomposition import PCA
pca= PCA(n_components= 25)    #or pca= PCA(25)
x = pca.fit_transform(x)
```

```python
x.shape
```

```python
# logisticRegression
```

```python
df1= pd.DataFrame(data=x)
df1
```

```python
df1.columns
```

```python
for n in range (0,25):
    cl= plt.boxplot(df1[n])
    plt.title("box plot to check outliers")
    plt.show(cl)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
t= le.fit_transform(df.iloc[:,-1])   # Label encoder at output data..
t
```

```python
df1['out'] = t
df1.head()
```

```python
from scipy.stats import zscore
z=np.abs(zscore(df1))
z
```

```python
threshold = 3
df1 = df1[(z<3).all(axis=1)]
```

```python
df1
```

```python
sns.countplot('out',data=df1)
plt.show()                 #we can see number of hit at mine and at rock
```

```python
s= df1.iloc[:,0:-1].values
u= df1.iloc[:, -1].values
```

```python
s.shape
```

```python
u.shape
```

```python
u.reshape(1,-1)
```

```python
from sklearn.model_selection import train_test_split
s_train,s_test,u_train,u_test=train_test_split(s, u, test_size = .33, random_state = 47)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(s_train,u_train)
lr.score(s_train,u_train)
```

```python
pred = lr.predict(s_test)
pred
```

```python
print(accuracy_score(u_test, pred))
print(confusion_matrix(u_test, pred))
print(classification_report(u_test, pred))
```

```python
# here we can see classification report and f1-score..
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
mod = [GaussianNB(), DecisionTreeClassifier(), KNeighborsClassifier(), SVC()]
for m in mod:
    m.fit(s_train,u_train)
    print(m,m.score(s_train,u_train))     # score at 0th then 1th, 2th ... position model(gaussianNB, Decisiontree...)
    pred=m.predict(s_test)
    print("\n accuracy score:", accuracy_score(u_test,pred))
    print("\n confusion matrix:", confusion_matrix(u_test,pred))
    print("\n classification_report:", confusion_matrix(u_test,pred))
  
```
