```python
from scipy.stats import zscore
z=np.abs(zscore(df1))
z
```

```python
threshold = 3
print(np.where(z>3))
```

```python
dfn=df1[(z<3).all(axis=1)]
```

```python
dfn
```

```python
x= dfn.iloc[:,0:-1]  # taking input data as head size 
x
```

```python
y=dfn.iloc[:,-1]  # taking as a target data.
y
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,_x_test,y_train,y_test=train_test_split(x,y, test_size=.33, random_state=47)
```

```python
lr=LinearRegression()    # creating instance for linear regression
```

```python
lr.fit(x_train,y_train)  #fitting the model with linear regression..
```

```python
lr.score(x_train,y_train)
```

```python
pred=lr.predict(_x_test)
pred
```

```python
print(mean_squared_error(y_test, pred))
print('r2score :', r2_score(y_test, pred))
```
