```python
from fuzzywuzzy import process
```

```python
new_array = []
```

```python
#df['Location'].value_counts()
```

```python
# creating a variable list and taking to 8 names in the list.
correct_names = ['Chennai', 'Banglore', 'Hyderabad', 'Mumbai', 'New Delhi', 'Kochi', 'Gurgaon', 'Secunderabad']

for i in df['Location']:
    x = process.extractOne(i, correct_names)
    if x[1] < 60:
        new_array.append('Unknown')
    else:
        new_array.append(x[0])
```

```python
x_train = train_final.drop(columns=['COST','RESTAURANT_ID'])
```

```python
y_train= train_final['COST']
```

```python
x_test = test_final.drop(columns =['RESTAURANT_ID'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
#prediting training model
pred = lr.predict(x_train)
```

```python
cvscore = cross_val_score(lr, x_train,y_train, cv=5, scoring='neg_mean_squared_error')
```

```python
print('cvscore:',cvscore)
print('mean:',cvscore.mean())
print('Stddev:',cvscore.std())
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(mean_squared_error(y_train,pred))
```
